### Column formating

The column 'Quantity (L)' is in object format but should be in float format. Let's convert it:

First, let's check the values appearing in the column:

```python
sales['Quantity (L)'].value_counts()
```

Let's try to infer data in another way

We are going to replace those '#' values by None

We can now replace the '#' by None:

```python
sales['Quantity (L)']=sales['Quantity (L)'].replace('#', np.nan)
```

We can now convert our object column into float:

```python
sales['Quantity (L)']=sales['Quantity (L)'].str.replace(',', '.').astype(float)
```

### Product Classification

In this part, we want to simplify the product categories that are complex.To do so, we create a new column called 'Classification' that takes the value 'AP' it the product is related to Aperitivi, and 'WH' if whisky:

```python
sales['Category'].unique()
```

We first define the lists of aperitives and whiskies:

```python
APs=['ABV', 'ABV SANS ALCOOL', 'AMARI CHINE FERNET', 'AMER', 'AMER SANS ALCOOL', 'APERITIF', 'APERITIFS', 'APERITIVI', 'BITTERGETRAENKE', 'COCKTAIL','COCKTAILS ET PUNCHS', 'COCKTAILS ET PUNCHS SANS ALCOOL', 'NON ALCO', 'RESTL MIX-GETRAENKE', 'SPARKLINGS', 'SPIRITS', 'TOTAL LIQUEURS & SPECIALITIES', 'TOTAL VERMOUTH', 'WEINHALTIGE GETRAENK', 'WERMUT']
WHs=['FLAVOURED WHISKY', 'TOTAL IMPORTED WHISKY', 'TOTAL SCOTCH WHISKY', 'WHISKIES', 'WHISKY']
```

Besides, we create a function to assign the good value of classification (AP for aperitives, and WH for Whiskies):

```python
def findClassification(row):
    if row['Category'] in APs:
        cat='AP'
    else:
        cat='WH'
    return cat
```

We can now apply it to our dataframe sales to create the new CLassification column:

```python
sales['Classification']=sales.apply(findClassification, axis=1)
```
